# Data Visualization 


```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

## Data table 
Health_heart_experiment

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)
df = load_dataset("./Health_heart_experimental.csv")
df.head(10)
```

## Line Chart
Age vs Systolic Blood Pressure

```python
data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36],
    "SysBP": [141, 109, 112, 106, 140, 115, 125, 111, 144, 113]
}

df = pd.DataFrame(data)

age = df["age"].tolist()
sysbp = df["SysBP"].tolist()

plt.plot(sysbp, age, color='blue', marker='o', linestyle='dashed')

plt.title("Age vs Systolic Blood Pressure")
plt.xlabel("Systolic Blood Pressure (SysBP)")
plt.ylabel("Age (Years)")

plt.show()
```

Age vs Diastolic Blood Pressure

```python
data = {
    "age": [64, 21, 30, 35, 39, 68, 50, 23, 34, 36],
    "DiaBP": [96, 100, 73, 90, 90, 96, 117, 85, 80, 109]
}

df = pd.DataFrame(data)

sns.lineplot(x='age', y='DiaBP', data=df)

plt.title("Age vs Diastolic Blood Pressure")
plt.xlabel("Age (Years)")
plt.ylabel("Diastolic Blood Pressure (DiaBP)")

plt.show()
```

## Bar Plot

Diastolic Blood Pressure by Age


```python

age = [21, 23, 30, 34, 35, 36, 39, 50, 64, 68]
diabp = [100, 85, 73, 80, 90, 109, 90, 117, 96, 96]

plt.bar(range(len(age)), diabp, color='skyblue')

plt.title("Diastolic Blood Pressure by Age")
plt.ylabel("Diastolic Blood Pressure (DiaBP)")
plt.xlabel("Individuals (Sorted by Age)")

plt.xticks(range(len(age)), age)

plt.show()
```

Countplot of Sex

```python

data = {
    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1],
}
df = pd.DataFrame(data)

sns.countplot(data=df, x='sex')
plt.title('Countplot of Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

#sex_counts = df['sex'].value_counts()
#print(sex_counts)
```

Average Age by Weight and Sex

```python
data = {
    'age': [64, 21, 30, 35, 39, 68, 50, 23, 34],
    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1],
    'weightKg': [69, 48, 69, 45, 92, 53, 56, 55, 67],
}
df = pd.DataFrame(data)

sns.barplot(x='weightKg', y='age', hue='sex', data=df)
plt.title('Average Age by Weight and Sex')
plt.xlabel('Weight (kg)')
plt.ylabel('Average Age')
plt.show()
```

## Pie Chart
proportion of Sex


```python
data = {
    
    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1]
}
df = pd.DataFrame(data)

df['sex'].value_counts().plot(kind='pie', autopct='%0.2f', labels=['Male', 'Female'],colors=['lightblue', 'pink'])
plt.title('Proportion of Sex')
plt.show()
```

## Box Plot
Box Plot of Age 

```python
data = {'age': [64, 21, 30, 35, 39, 68, 50, 23, 34]}
df = pd.DataFrame(data)

sns.boxplot(df['age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.ylabel("")
plt.show()
```

Age Distribution by Sex and Indication

```python

data = {
    'age': [64, 21, 30, 35, 39, 68, 50, 23, 34],
    'sex': [1, 1, 0, 1, 0, 1, 1, 1, 1],
    'indication': [1, 0, 0, 0, 1, 0, 0, 0, 0]
}
df = pd.DataFrame(data)

df['sex'] = df['sex'].map({1: 'Male', 0: 'Female'})
df['indication'] = df['indication'].map({1: 'Yes', 0: 'No'})

sns.boxplot(x='sex', y='age', hue='indication', data=df)
plt.title('Age Distribution by Sex and Indication')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

## Histogram
Histogram of Age

```python

data = {'age': [64, 21, 30, 35, 39, 68, 50, 23, 34]}
df = pd.DataFrame(data)


plt.hist(df['age'], color='skyblue', edgecolor='skyblue')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Distribution of Age with KDE (Kernel Density Estimation):

```python

data = {'age': [64, 21, 30, 35, 39, 68, 50, 23, 34]}
df = pd.DataFrame(data)

sns.displot(df['age'], kind="hist", kde=True, color='skyblue')
plt.title('Distribution of Age with KDE')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Age Distribution by Indication :

```python

data = {
    'age': [64, 21, 30, 35, 39, 68, 50, 23, 34],
    'indication': [1, 0, 0, 0, 1, 0, 0, 0, 0]
}
df = pd.DataFrame(data)

sns.kdeplot(df[df['indication'] == 0]['age'], label='No Indication', color='red')
sns.kdeplot(df[df['indication'] == 1]['age'], label='Indication', color='blue')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
